```python


import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")

display(data)
```

```python
%matplotlib inline
```

```python
data.plot(x="Rank", y=["P25th", "Median", "P75th"])
```

```python
data.plot(x="Rank", y=["P25th", "Median", "P75th"], kind="hist", bins=25, grid=False, figsize=(12,8), color=[ 'Yellow', 'Black', 'Red'], zorder=2, rwidth=0.9)
```

```python
data.plot(x="Major", y=["P25th", "Median", "P75th"], kind="box", rot=5, fontsize=8)
```
